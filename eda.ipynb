```python
import pandas as pd 
import numpy as np
import os
```

```python
# Displaying df with all columns available 
pd.set_option("display.max_columns",None)
```

```python
passing_air_yards_df = pd.read_csv("Data/Adv_Stats/Passing_Air_Yards_2018.csv")
passing_accuracy_df = pd.read_csv("Data/Adv_Stats/Passing_Accuracy_2018.csv") \
                                    .drop(columns= ["G","Cmp","Att","Yds"])
passing_pressure_df = pd.read_csv("Data/Adv_Stats/Passing_Pressure_2018.csv") \
                                    .drop(columns=["G","Cmp","Att","Yds","PktTime"])
passing_2018_df = pd.merge(passing_air_yards_df,passing_accuracy_df,how='inner',on="Tm").merge(passing_pressure_df,how='inner',on="Tm")
```
