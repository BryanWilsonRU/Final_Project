```python
import pandas as pd
```

```python
offense_2005 = pd.read_csv("../Team_Stats/2005_team_offense.csv")
defense_2005 = pd.read_csv("../Team_Stats/2005_team_defense.csv")
offense_2006 = pd.read_csv("../Team_Stats/2006_team_offense.csv")
defense_2006 = pd.read_csv("../Team_Stats/2006_team_defense.csv")
offense_2007 = pd.read_csv("../Team_Stats/2007_team_offense.csv")
defense_2007 = pd.read_csv("../Team_Stats/2007_team_defense.csv")
offense_2008 = pd.read_csv("../Team_Stats/2008_team_offense.csv")
defense_2008 = pd.read_csv("../Team_Stats/2008_team_defense.csv")
offense_2009 = pd.read_csv("../Team_Stats/2009_team_offense.csv")
defense_2009 = pd.read_csv("../Team_Stats/2009_team_defense.csv")
```

# 2005

```python
df_2005 = offense_2005.merge(defense_2005, how='inner', left_on='Tm_Off', right_on='Tm_Def')
df_2005.head()
```

```python
df_2005 = df_2005.drop('Year_Def', 1)
df_2005 = df_2005.drop('Win_Def', 1)
df_2005 = df_2005.drop('Loss_Def', 1) 
df_2005 = df_2005.drop('Tm_Def', 1)
df_2005 = df_2005.drop('G_Def', 1)
df_2005.head()
```

```python
df_2005 = df_2005.rename(columns={'Year_Off': 'Year', 'Win_Off': 'Win', 'Loss_Off': 'Loss', 'Tm_Off': 'Team', 'G_Off': 'Games'})
df_2005.head()
```
