```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
OFF10 = "2010OffenseRenamed.csv"
OFF11 = "2011OffenseRenamed.csv"
OFF12 = "2012OffenseRenamed.csv"
OFF13 = "2013OffenseRenamed.csv"
OFF14 = "2014OffenseRenamed.csv"
DEF10 = "2010DefenseRenamed.csv"
DEF11 = "2011DefenseRenamed.csv"
DEF12 = "2012DefenseRenamed.csv"
DEF13 = "2013DefenseRenamed.csv"
DEF14 = "2014DefenseRenamed.csv"
```

```python
# Import & read csv
OFF10df = pd.read_csv(OFF10)
OFF11df = pd.read_csv(OFF11)
OFF12df = pd.read_csv(OFF12)
OFF13df = pd.read_csv(OFF13)
OFF14df = pd.read_csv(OFF14)
DEF10df = pd.read_csv(DEF10)
DEF11df = pd.read_csv(DEF11)
DEF12df = pd.read_csv(DEF12)
DEF13df = pd.read_csv(DEF13)
DEF14df = pd.read_csv(DEF14)
```

```python
df_2010 = OFF10df.merge(DEF10df, how='inner', left_on='Tm_Off', right_on='Tm_Def')
df_2010.head()
```

```python
df_2010 = df_2010.drop('Year_Def', 1)
df_2010 = df_2010.drop('Win_Def', 1)
df_2010 = df_2010.drop('Loss_Def', 1) 
df_2010 = df_2010.drop('Tm_Def', 1)
df_2010 = df_2010.drop('G_Def', 1)
```

```python
df_2010 = df_2010.rename(columns={'Year_Off': 'Year', 'Win_Off': 'Win', 'Loss_Off': 'Loss', 'Tm_Off': 'Team', 'G_Off': 'Games'})
df_2010
```
