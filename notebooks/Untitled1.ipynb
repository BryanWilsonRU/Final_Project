```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Files to Load
stats_path = "../Data/Team_Stats/2000-2019_cleaned_data5.csv"
```

```python
# Read files and store into Pandas data frames
stats_df = pd.read_csv(stats_path)
```

```python
# Fill NaN values with 0
stats_df = stats_df.fillna(0)
```

```python
# Display sample DF for review
stats_df.head()
```

```python
# Calculate Differences between prediction and actual
stats_df['PT_Pct_Diff'] = stats_df['NY_Win/Loss_Pct'] - stats_df['PT_Result']
stats_df['Pct_Diff_Squared'] = stats_df['PT_Pct_Diff'] * stats_df['PT_Pct_Diff']
stats_df['PT_Win_Diff'] = stats_df['NY_Win'] - stats_df['PT_Wins']
stats_df['Win_Diff_Squared'] = stats_df['PT_Win_Diff'] * stats_df['PT_Win_Diff']
stats_df['PT_Loss_Diff'] = stats_df['NY_Loss'] - stats_df['PT_Losses']
stats_df['Loss_Diff_Squared'] = stats_df['PT_Loss_Diff'] * stats_df['PT_Loss_Diff']
stats_df.head()
```

```python
#Write combined DataFrame to .csv
stats_df.to_csv("../Data/Team_Stats/2000-2019_cleaned_data_Tableau.csv",index=False)
```
