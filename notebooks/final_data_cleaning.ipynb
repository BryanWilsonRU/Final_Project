```python
import pandas as pd
```

```python
df_2000_2004 = pd.read_csv("../Data/Team_Stats/2000-2004_cleaned_data.csv")
df_2005_2009 = pd.read_csv("../Data/Team_Stats/2005-2009_cleaned_data.csv")
df_2010_2014 = pd.read_csv("../Data/Team_Stats/2010-2014_cleaned_data.csv")
df_2015_2019 = pd.read_csv("../Data/Team_Stats/2015-2019_cleaned_data.csv")
```

```python
combined_df = df_2000_2004[:158].append([df_2005_2009[:160], df_2010_2014[:160], df_2015_2019[:160]])
combined_df
```

```python
combined_df['Overall Production'] = ""
combined_df
```

```python
for index, row in combined_df.iterrows():
    off_yards = combined_df['Total Yds_Off']
    def_yards = combined_df['Total Yds_Def']
    
    combined_df.loc[index, "Overall Production"] = off_yards - def_yards
    
```
