```python
import pandas as pd
```

```python
#All defense csvs
df15=pd.read_csv('../Data/Team_Stats/2015_team_defense.csv')
df16=pd.read_csv('../Data/Team_Stats/2016_team_defense.csv')
df17=pd.read_csv('../Data/Team_Stats/2017_team_defense.csv')
df18=pd.read_csv('../Data/Team_Stats/2018_team_defense.csv')
df19=pd.read_csv('../Data/Team_Stats/2019_team_defense.csv')
# df_merge= df15[:32].append([df16[:32], df17[:32], df18[:32], df19[:32]])
# df_merge.to_csv('2015-2019_defense.csv')
```

```python
#All offense csvs
of15=pd.read_csv('../Data/Team_Stats/2015_team_offense.csv')
of16=pd.read_csv('../Data/Team_Stats/2016_team_offense.csv')
of17=pd.read_csv('../Data/Team_Stats/2017_team_offense.csv')
of18=pd.read_csv('../Data/Team_Stats/2018_team_offense.csv')
of19=pd.read_csv('../Data/Team_Stats/2019_team_offense.csv')
# of_merge= of15[:32].append([of16[:32], of17[:32], of18[:32], of19[:32]])
# of_merge.to_csv('2015-2019_offense.csv')
```

```python
#2015
df_2015 = of15.merge(df15, how='inner', left_on='Tm_Off', right_on='Tm_Def')
df_2015.head()
```

```python
df_2015 = df_2015.drop('Year_Def', 1)
df_2015 = df_2015.drop('Win_Def', 1)
df_2015 = df_2015.drop('Loss_Def', 1) 
df_2015 = df_2015.drop('Tm_Def', 1)
df_2015 = df_2015.drop('G_Def', 1)
df_2015.head()
```

```python
df_2015 = df_2015.rename(columns={'Year_Off': 'Year', 'Win_Off': 'Win', 'Loss_Off': 'Loss', 'Tm_Off': 'Team', 'G_Off': 'Games'})
df_2015.head()
```
